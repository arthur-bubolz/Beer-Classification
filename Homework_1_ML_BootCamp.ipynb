```python
#Question 7:
my_value = 'ISLAND'
results = df.loc[df['ocean_proximity'] == my_value]
x = results[['housing_median_age','total_rooms','total_bedrooms']]
```

```python
x.head()
```

```python
x_transpose = x.T

xtx = np.dot(x_transpose, x)
```

```python
#computing the inverse of xtx:

m_inverse = np.linalg.inv(xtx)

print(m_inverse)
```

```python
y = [950, 1300, 800, 1000, 1300]

first = np.dot(m_inverse, x_transpose)
w = np.dot(first, y)

print(w)
```
